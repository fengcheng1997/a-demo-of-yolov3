```python
import re
import os
import shutil
import urllib.request

# delete the original pictures
shutil.rmtree("imgs_of_POI")
os.mkdir("imgs_of_POI")

# download pictures
def craw(url, page):
    html1=urllib.request.urlopen(url).read()
    html1=str(html1)
#     print(html1)
    pat1='<div id="content">.+?<div class="paginator">'
    result1=re.compile(pat1).findall(html1)
    print(type(result1), len(result1))
    print("\n")
    result1=str(result1)
    pat2='<img src="(https://.+?.jpg)" />'
    print(type(pat2), type(result1))
    #img width="220" height="282" class="err-product" data-img="1" source-data-lazy-img
    imagelist=re.compile(pat2).findall(result1)
#     print("data:",imagelist)
    print(len(imagelist))
    x=1
    for imageurl in imagelist:
        imagename = "./imgs_of_POI/"+str(page)+'-'+str(x)+".jpg" #前面是文件夹，后面是图片名字
        # imageurl = "http://"+ imageurl
        # print(imageurl, imagename)
        try:
            urllib.request.urlretrieve(imageurl,filename=imagename)
        except urllib.error.URLError as e:
                x+=1
        x+=1
```
